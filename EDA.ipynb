```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
sns.pairplot(car)
```

```python
#Ploting three different values on the same graph
sns.relplot(x='Year', y='Mileage', hue='Price', data=car)
```
